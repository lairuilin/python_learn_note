```python
book = []

with open('bucket_utf8.txt', 'r') as f:    ###讀檔 一行一行讀 類似java的buffer reader
    for line in f.readlines():
        line = line.rstrip()          ##刪除空白字元
        book.append(line.split('\t'))  ##將一行讀入的值以tab分隔，並加入list內
```

```python
from collections import Counter  ##為一個容器用的package，包含了許多容器，類似JAVA的container or C++的vector
from itertools import chain   ##較高速的迭代工具
```

```python
book_count = dict(Counter(chain.from_iterable(book)))  ##計算書本被讀過幾次 https://www.cnblogs.com/zhenwei66/p/6593395.html
## Counter可以用來追蹤值的出現次數
```

```python
len(book_count.keys())  
```

```python
book_new = []
```

```python
for key, values in book_count.items():  ##key=dict的key values等於書看過的次數
    if values >= 1 and values <= 26:
        book_new.append(key)
```

```python
len(book_new)
```

```python
book_new = set(book_new)   ##得到>1本小於26本的書籍集合
book_fix = list(map(lambda x: list(filter(lambda y: y in book_new, x)), book))  ##filter過濾掉不符合條件的
### lambda是運算式! 所以filter(lambda y: y in book_new, x)意味過濾
###filter(function, iterable)  list將結果轉為list map將每一個元素傳入list 但是同時會返回每一個調用結果
```

```python
with open('bucket_utf8_1_26.txt', 'w') as f:
    for line in book_fix:
        f.write('\t'.join(line)+'\n')
```

```python
from openpyxl import Workbook
import pymysql
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("是否借閱書籍矩陣(書名版).csv")
```

```python
df1=df1.iloc[:,1:]   ##去掉index
print(df1)  
df1=df1.loc[:, (df1 != 0).any(axis=0)]  ## 把dataframe的col部分 不等於0的col取出並重新指定入df1
print(df1)
df1= df1.loc[:, (df1 != 1).any(axis=0)]   ##把dataframe的col部分 不等於1的col取出並重新指定入df1
df1 = df1.loc[:,df1[df1>0].count() > 14 ]  ##將整個dataframe中col>14的取出並存入
df1 = df1.loc[:,df1[df1>0].count() < 26 ] 
print(df1)
```

```python

frequent_books = apriori(df1 ,min_support=0.3 ,use_colnames=1)
rules_books = association_rules(frequent_books,metric='lift',min_threshold=1.25) ##min_threshold是搭配metric用的
# print(frequent_books)
frequent_books.to_csv("借閱種類關聯.csv",encoding='utf_8_sig')
print(rules_books)
rules_books.to_csv("借閱種類lift數值.csv",encoding='utf_8_sig')
```
